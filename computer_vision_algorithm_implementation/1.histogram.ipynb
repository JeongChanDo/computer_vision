```python
import math
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("./res/lena.jpg")

print(img)
```

```python
#change to graysclae
img = img.convert("L")
print(img)
```

```python
#change to np array
img = np.asarray(img)
print(type(img))
print(img)
```

```python
#show gray scale image
plt.imshow(img,cmap="gray")
```

```python
def Histogram(img):
    row, col = img.shape
    hist = np.zeros(256)
    for i in range(0, row):
        for j in range(0, col):
            hist[img[i, j]] += 1
    
    return hist
```

```python
hist = Histogram(img)
print(hist[:100])

plt.bar(np.arange(0,256), hist, width=5)
```
